Классификация/регрессия (необязательный)

Задачу построения рекомендаций можно свести к классификации или регрессии.

Мы можем подавать на вход в нейронную сеть последовательность покупок пользователя. И параллельно на другой вход в нейронную сеть подавать новый продукт / продукты, который мы собираемся этому пользователю предложить. Если новый продукт есть в следующей покупке пользователя, то целевая функция равна 1. Если новый продукт отсутствует в следующей покупке пользователя, то целевая функция равна 0.

В нейронной сети можно использовать:

Эмбеддинги
Сверточные слои
Рекуррентные слои
LSTM / GRU
скалярное произведение тензоров
конкатенацию
dropout
etc


Построение рекомендаций

После того как нейронная сеть будет натренирована, мы сможем для каждого пользователя подать на вход этой нейронной сети историю покупок пользователя. На другой вход подать новый продукт, который собираемся предложить. А результат работы модели нам даст возможность ранжировать продукты по их релевантности для каждого человека.

Проблема таких рекомендаций:

Каждый вариант выбора необходимо скорить отдельно
Матрицу предпочтений приходится заполнять по одной ячейке
 Оказывается очень долго
