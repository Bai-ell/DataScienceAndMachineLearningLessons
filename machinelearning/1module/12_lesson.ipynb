Jupyter

Jupyter Notebook — среда разработки, удобный инструмент для создания аналитических отчётов, позволяет добавлять код, изображения, комментарии, формулы и графики.

Режимы работы с ячейками:

режим правки (ввода) — редактирование содержимого, копирование, вставка, есть возможность возращаться к ячейкам и оперативно вносить правки;
командный режим — выполнение команд, результат выводится отдельно по каждой команде.
По умолчанию, notebook сохраняются самостоятельно через опредлённые промежутки времени, но можно дополнительно сохранять файлы самому. На видео демонстрируются полезные комбинации клавиш.
