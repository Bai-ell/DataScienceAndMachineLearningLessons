Примеры архитектур сверточных сетей VGG-16 и VGG-19 | #15 нейросети на Python

```python
import numpy as np 
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow import keras
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from io import BytesIO
from PIL import Image
from tensorflow.keras.preprocessing import image as kp_image
from tkinter import Tk
from tkinter.filedialog import askopenfilename
```

```python
model = keras.applications.VGG16()
```

```python
Tk().withdraw()  # Скрыть главное окно
filename = askopenfilename(title="Выберите файл")  # Выбор файла
print(f"Выбранный файл: {filename}")
```

```python
# Открыть изображение
img = Image.open(filename)
img = img.convert("RGB")
# Привести изображение к размеру (224, 224)
img = img.resize((224, 224))
# Преобразовать изображение в формат numpy array
img_array = np.array(img)

# Показать изображение через matplotlib
plt.imshow(img_array)
plt.axis('off')  # Отключить оси
plt.show()
```

```python
x = keras.applications.vgg16.preprocess_input(img_array)
print(x.shape)
x = np.expand_dims(x , axis=0)
print(x.shape)
```

```python
res = model.predict(x)
print(np.argmax(res))
```
