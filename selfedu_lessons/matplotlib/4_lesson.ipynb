```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import numpy as np 
%matplotlib inline
```

```python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set(xlim=(-5, 30), ylim=(-1, 6))


plt.show()
```

```python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set(xlim=(-5, 30),)
ax.set(ylim=(-1, 6))




plt.show()
```

```python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set_xlim(xmin=-5, xmax=30)
ax.set_ylim(-1, 6)





plt.show()
```
