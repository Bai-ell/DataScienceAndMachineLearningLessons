```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import numpy as np 
%matplotlib inline
```

```python
x = np.array([1, 1, 5, 5, 1])
y = np.array([1, 5, 5, 1, 1])
plt.plot(x, y)
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a*a for a in y])


y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]


# plt.plot(x, y, '--')
# plt.plot(x2, y2, ':')

lines = plt.plot(x, y,'s--r', x2, y2, 'o:m')
plt.grid()
plt.show()
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.fill_between(x, y, where = (y < 0), color = 'red', alpha = 0.5)
plt.fill_between(x, y, where = (y > 0), color = 'g', alpha = 0.5)


plt.grid()
plt.show()
```
