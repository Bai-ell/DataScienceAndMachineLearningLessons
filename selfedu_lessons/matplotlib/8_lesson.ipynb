```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import numpy as np 
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
ax.plot(np.arange(0, 5, 0.25), '--',label = 'line1')
ax.plot(np.arange(0, 10, 0.5), ':s',label = 'line2')

ax.legend(['1', '2'], loc='upper right')

plt.show()
```
