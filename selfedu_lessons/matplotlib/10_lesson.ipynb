```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import numpy as np 
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
ax.hist(y,  50)
ax.grid()



plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

# y = np.random.normal(0, 2, 500)
# ax.hist(y,  50)

x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(1, 5, len(x))
ax.bar(x, y)
ax.grid()



plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
x = np.linspace(np.min(y), np.max(y), 10)
bars = [len(y[np.bitwise_and(y >= x[i], y < x[i+1])]) for i in range(len(x)-1)]

# ax.hist(y,  50)

# x = [f'H{i+1}' for i in range(10)]
# y = np.random.randint(1, 5, len(x))
ax.bar(range(len(bars)), bars)
ax.grid()



plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()



x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(-20, 20, len(x))



ax.bar(x, y, width=0.5,  linewidth = 2, edgecolor='r', yerr = 2, bottom= 10)
ax.grid()



plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()



x = np.arange(10)
y1 = np.random.randint(3, 20, len(x))
y2 = np.random.randint(3, 20, len(x))
w = 0.3
ax.bar(x, y1, width=w, label='y1', color='b')
ax.bar(x + w, y2, width=w, label='y2', color='r')


ax.bar(x, y, width=0.5,  linewidth = 2, edgecolor='r', yerr = 2, bottom= 10)
ax.grid()



plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()



x = np.arange(10)
y1 = np.random.randint(3, 20, len(x))
y2 = np.random.randint(3, 20, len(x))
w = 0.3
ax.bar(x - w/2, y1, width = w)
ax.bar(x - w/2, y2, width = w)


# ax.bar(x, y, width=0.5,  linewidth = 2, edgecolor='r', yerr = 2, bottom= 10)
ax.grid()



plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()



vals = [10, 40, 23, 30, 7]
labes = ['toyota', 'honda', 'ford', 'mercedes', 'audi']
ax.pie(vals, labels=labes)


ax.grid()



plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()



vals = [10, 40, 23, 30, 7]
labes = ['toyota', 'honda', 'ford', 'mercedes', 'audi']
exp = [0.1, 0.2, 0.4, 0, 0.5]
ax.pie(vals, labels=labes, autopct='%1.1f%%', startangle=90, explode=exp,  wedgeprops=dict(width=0.5), shadow=True)


ax.grid()



plt.show()
```
