```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import numpy as np 
```

```python
y = np.array([1, 2, 3, 4, 5])
x = np.array([a*a for a in y])
y2 = [ 0, 2 , 3, 4, 5, 6, 7]
x2 = [ i +1 for i in y2]

lines = plt.plot(x, y, x2, y2)


plt.minorticks_on()
plt.grid(which= 'major', color = '#444', linewidth = 2)
plt.grid(which= 'major', color = '#aaa', ls = ':')


plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
plt.figtext(0.04, 0.6, 'Text in window state')
fig.suptitle('Tittle of the window')
ax.set_xlabel('Ox')
ax.set_ylabel('Oy')

ax.text(0.5, 0.5, 'Random text in coordinates')
ax.annotate('Annotation', xy=(0.2, 0.4), xytext=(0.6, 0.7),
            arrowprops=dict(facecolor='gray', shrink=0.1))
```
