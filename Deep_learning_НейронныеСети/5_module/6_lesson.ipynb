Pooling слой будет сжимать изображение в два раза. На выходе получим небольшой тензор, который можно вытянуть в вектор, — это и будет представлением для картинки. Это представление будем называть embedding или результатом encoder'а.

Как развернуть полученный вектор из числа обратно в картинку? Кажется естественным повторить преобразование encoder'а в обратную сторону. Как обратить результат pooling слоя? Используем Un-pooling слой.
