Где используется детектирование объектов?


Приложений детектирования объектов на самом деле очень много, так же как и в семантической сегментации.

Назовём несколько из них:

детектирование лиц и людей (в системах видеонаблюдения, магазинах и других местах массового скопления людей и др.);
детектирование продуктов (автоматизированный просмотр продуктов на полках магазинов, «умный» холодильник и др.);
автономные транспортные средства.
