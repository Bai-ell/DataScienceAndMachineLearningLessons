Метод R-CNN

R-CNN (region convolutional neural network ) — нейросетевой метод для детектирования объектов .

Разберем, как работает метод R-CNN:

Методы для определения region proposals по заданной картинке дают набор потенциальных боксов. Задача — всё это подмножество боксов просмотреть. Например, пусть алгоритм selective search выдал 1000 потенциальных боксов. Каждый бокс, т.е. каждый регион на картинке, соответствующий этому боксу, преобразуется к фиксированному размеру.

Берем каждый регион, прогоняем через нейронную сеть, получаем классификацию. Это похоже на скользящее окно, но разница в том, что с помощью алгоритма region proposals предварительно было ограничено количество потенциальных регионов для классификации.

Можно улучшить этот метод, в случае если region proposals немного ошибся и не точно предсказал нам границы объекта. Для этого навешивается некоторая регрессия, благодаря чему на выходе из одного предсказания будет не просто распределение по классам, но ещё и регрессия на bounding box. Но так как уже изначально region proposals  содержал предсказание, где находится bounding box, необязательно предсказывать совершенно новое значение bounding box в этой регрессии, а можно предсказать лишь его смещение. Смещение покажет, как можно поправить bounding box на Δx или Δy, сместить координаты центра, ширину или длину, чтобы bounding box был более точным.

Проблемы R-CNN

Нельзя обучить алгоритм region proposals из первой части метода, который к тому же может быть потенциально медленным.
Проблема скользящего окна: свёрточная нейронная сеть запускается несколько раз для разных регионов, что тоже не очень эффективно.
Решить вторую проблему можно, воспользовавшись FAST R-CNN — модификацией метода R-CNN.
