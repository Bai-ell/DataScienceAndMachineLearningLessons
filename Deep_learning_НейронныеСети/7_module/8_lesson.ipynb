Дилатационная свёртка

Как работает дилатационная свёртка?

Суть дилатационной свёртки в том, чтобы  увеличить размер рецептивного поля на входной картинке, чтобы нейронная сеть захватывала и анализировала большую часть входной картинки для предсказания ответа на каком-то выходном пикселе.

В дилатационной свёртке скользящее окно считывает значение входной картинки с некоторым разрывом, т.е. количество параметров остается прежним, и расстояние между точками, которые мы семплируем  на входном изображении, можно регулировать.
