Классическая сегментация

Алгоритмы классической сегментации

Суперпиксели реализуют разбиение изображения на множество мелких фрагментов, представляющих из себя относительно однородные группы расположенных рядом пикселей.  Каждый суперпиксель потенциально является атомарным фрагментом изображения, т.е. все входящие в него пиксели рассматриваются при дальнейшей обработке как единое целое.

Условные случайные поля (англ. conditional random field, CRF) — это статистический метод классификации, основанный на графах. Характерным отличием этого метода является возможность учитывать «контекст» классифицируемого объекта. CRF можно использовать в сочетании с суперпикселями или нейросетевыми подходами. Одним из главных достоинств этой модели является то, что она не требует моделировать вероятностные зависимости между наблюдаемыми переменными.

Скрытые значения в CRF — это искомые метки классов для каждого пикселя.

Наблюдаемые значения — это унарный потенциал, т.е. параметр, который мы знаем для каждого пикселя по нашей задаче, например, цвет. 

Связи между элементами — это парный потенциал, который характеризуется функцией «энергия» (эта функция представляется в виде суммы потенциальных функций, называемых также потенциалами на кликах), которая позволяет выразить зависимости между соседними элементами случайного поля.
