 Сегментация объектов. MASK R-CNN

Задача сегментации очень похожа на детектирование объектов: тоже необходимо найти несколько объектов и проклассифицировать их, но кроме ограничивающего прямоугольника необходимо найти карту сегментации.

Возникает идея соединить семантическую сегментацию и детектирование объектов, которая воплощается в методе MASK R-CNN.

Идея метода:

берем FASTER R-CNN за основу детектирования объектов;
после ROI Pooling накручиваем часть сегментационной сети, ответственную за декодирование.
