Обучение локализации

Как обучать такую нейронную сеть?

Помним: чтобы обучить классификатор, мы стандартно используем функцию Softmax, потом Cross Entropy в качестве функции потерь и всё это дело обучаем.

И можем этот самый Loss навесить на вектор, выходной из нашей нейронной сети, который соответствует классификации. Если у нас один вектор на выходе, который соответствует и классификации и регрессии одновременно, мы берём из него только часть и подаём в классификационный Loss (Cross Entropy).

А то, что касается bounding box, мы подаем на другой Loss, которой ответственен за регрессию. Как правило, там используется опять же евклидово расстояние. Мы «тянем» наш предсказанный bounding box к известному bounding box, который соответствует этому классу. А все остальные bounding box, если у нас их много, можем тянуть к каким-то неправильным значениям, которые не присутствуют на картинке. Это делается для того, чтобы дать сети понять, что с этим bounding box ничего вообще делать не нужно.

В итоге мы получили два Loss'a, а наш финальный Loss будет их суммой. Но так как у нас в задаче два Loss'a различной природы, правильно будет не просто суммировать их, а добавить коэффициент λ.
