Как улучшить FCN?

В случае обучения нейросети сегментации мы к входной картинке добавляем разметку, т.е. указываем, к какому классу принадлежит каждый пиксель:

применяем FCN;
получаем предсказание;
применяем функцию argmax;
используем распределение и Ground-Truth;
применяем функцию Loss для каждого пикселя;
суммируем по всем пикселям изображения.

Операция upscale не даёт возможности увидеть мелкие, высокочастотные детали. Для решения проблемы качества можно улучшить pooling (pooling — это не единственный способ понижать размерность изображений).

Вместо pooling слоя можно применить свёртку со страйдом. Она даёт возможность понизить размерность картинки, например, увеличив шаг скользящего окна в 2 пикселя. Таким образом, каждое второе окно будет соответствовать какому-то выходному пикселю и в выходном окне получится в два раза меньше пикселей.
