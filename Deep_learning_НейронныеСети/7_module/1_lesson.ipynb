В этом модуле мы поговорим о семантической сегментации.

Сегментация — это разбиение изображения на непересекающиеся области, покрывающие все изображение, и однородные по некоторым признакам: цвет, яркость, текстура и т. д.

Задачи компьютерного зрения:

классификация (Label) — присваивание ярлыка целому изображению; 
локализация (Bounding box + Label) — определение рамки вокруг объекта и его описание;
локализация (Heatmap) — определение объекта в виде тепловой карты;
детектирование объектов — создание нескольких рамок на изображении;
семантическая сегментация — создание точных сегментов, содержащих объекты, на изображении;
сегментация объектов — создание точных сегментов, где каждый объект в рамках одного класса разделяем.
В семантической сегментации важно классифицировать не изображение, а каждый пиксель картинки, т.е. выделить некоторые сегменты картинки и каждому из сегментов присвоить метку класса.
