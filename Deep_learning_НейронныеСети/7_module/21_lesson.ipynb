 Метод SSD

Идея SSD пошагово:

прогоняем картинку через глубокую нейронную свёрточную сеть CNN, получая карты признаков на разных уровнях этой сети, которые позволяют захватывать объекты различной сложности;
к каждой карте признаков применяем детектор, предсказывающий регионы с метками классов;
чтобы собрать воедино множество регионов, полученных на втором шаге, используем алгоритм Non-Maximum Suppression, группирующий боксы и выдающий один усредненный с соответствующим классом.
