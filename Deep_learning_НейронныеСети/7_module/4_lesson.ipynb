Классификация изображений

Скользящее окно — прямоугольник, который перемещается слева направо и сверху вниз по изображению с некоторым шагом по осям. Из частей изображений, попавших в это окно, выделяются некоторые признаки, и проводится их классификация. Результаты такой классификации затем используются для определения положения искомого объекта или объектов.

Тепловая карта (heat map) — это способ графического представления данных, при котором значения сгруппированы в матрицу и отображаются с помощью цвета.

Маскирование — это способ очистить слишком плотное или детализированное содержимое карты, используя объекты слоя маскирования, или блокировать объекты другого слоя в местах наложений. Существует два способа маскирования:

Маскирование слоем — все объекты одного слоя маскируют перекрывающиеся объекты другого слоя на карте или сцене.
Маскирование на уровне объектов — маскирование задаётся пообъектно, как указано классом отношений между двумя слоями.
Одна из проблем такого подхода — неоптимальность, т.к. обученные таким образом нейросети работают с фиксированным размером изображения. В случае увеличения размеров входного изображения свёрточные слои его хорошо обработают, а полносвязные слои нет.

Решение проблемы неоптимальности

При помощи свёртки 1 × 1 можно решить данную проблему, заменив полносвязные слои. Для примера рассмотрим алгоритм применения свертки 1 × 1 в нескольких ситуациях:

1. Выход  последнего свёрточного слоя в пространственных измерениях имеет размер 1 × 1 и n количество каналов.  Если бы эта сеть была с обычным полносвязными слоями, действия были бы следующими: 

векторизовать трёхмерных тензор;
к полученному вектору применить полносвязные слои;
на выходе сети перевести все в трёхмерный формат. Например, для тепловой карты мы бы вернулись к канальному измерению, поместив k-классов в третье измерение, а пространственное измерение оставив 1 × 1 пиксель, и поместили бы этот пиксель на тепловую карту.

2. Выход  последнего свёрточного слоя в пространственных измерениях имеет размер 1 × 1 и n карт признаков. Если мы применим к входу свёртку в пространственном измерении 1 × 1, но у которой есть m выходных каналов, это действие будет эквивалентно матричному умножению. То же действие мы можем применить ко второму слою: применить свертку 1 × 1 и получить новые карты признаков, при этом пространственное измерение не изменится, а канальное изменится, т.к. в свёрточных слоях мы можем задавать выходные канальные измерения. Нет необходимости проводить обратное преобразование,  т.к. мы сразу получаем финальный трёхмерный тензор с пространственным измерением 1 × 1 и двумя выходными каналами. Если мы в этот момент подадим на вход картинку большего размера, то и выход последнего свёрточного слоя станет больше и проблемы с применением полносвязного слоя не будет.

3. Выход последнего свёрточного слоя исходной сети имеет размеры больше 1 × 1. В этом случае используем в первой свёртке ядро не 1х1, а ядро того же размера, что и выход.

Решение проблемы слабой локализации

Для решения проблемы слабой локализации можно применить функцию  argmax (или arg max, аргумент максимизации — значение аргумента, при котором данное выражение достигает максимума) — в каждом пикселе вычислить индекс того класса, где достигается максимальное значение.

После или до применения функции argmax, можно применить upscale — увеличение размера изображения с сохранением пропорций, и мы получим карту семантической сегментации (но она и будет не вполне точной).
