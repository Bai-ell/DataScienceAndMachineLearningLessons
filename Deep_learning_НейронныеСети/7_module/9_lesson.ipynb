Детектирование объектов. Введение

Задачи в компьютерном зрении

В этом модуле мы поговорим о детектировании объектов.

Но прежде чем перейти к этой теме, вспомним, какие ещё есть задачи в компьютерном зрении, чтобы не путать детектирование, сегментацию, локализацию и т.д.

Разобьём все задачи (методы) на две группы:

1 группа: на картинке представлен один объект. В этом случае самое простое, что можно сделать, — классификация. Такая задача («на входе картинка — на выходе метка класса») стоит во главе угла многих других задач компьютерного зрения. Локализовать здесь можно двумя способами: с помощью ограничивающего прямоугольника (bounding box или просто бокс), либо сделав локализацию в виде тепловой карты.

2 группа: на картинке может присутствовать несколько объектов. Необходимо эти объекты найти и тоже проклассифицировать. Один из способов это сделать — для каждого объекта найти ограничивающий прямоугольник (бокс) и присвоить ему метку класса. Другие способы связаны с сегментацией: например, можно каждый пиксель проклассифицировать, найти сегменты принадлежащие разным классам, не разделяя данные одного и того же класса. И третье, самое сложное, можно сделать детектирование объектов и сегментацию в одном флаконе, разделяя объекты внутри одного класса и для каждого объекта кроме примерного позиционирования определить точную маску сегментация.

В этом модуле мы поговорим о трех оставшихся задачах, которые мы еще не рассмотрели:

классификация + локализация;
детектирование объектов;
сегментация объектов.
Во главу угла поставим детектирование объектов, но к нему придём через локализацию и посмотрим, как от него перейти к сегментации объектов.
