Приложения сегментации

Перечислим примеры практического применения семантической сегментации.

Семантическая сегментация применима в следующих сферах: 

автономные транспортные средства (сегментация изображений с камеры;  классификация пешехода на дороге, транспортного средства, знаков и пр.);
медицинские изображения (сегментация тканей, опухолей, аномалий и их характеристик (площадь, динамика));
спутниковая и аэрофотосъемка (распознавание алгоритмом географических объектов и их классификация по определенным признакам);
сегментация людей (деление объектов в сцене на сегменты для дальнейшей работы над ними).
