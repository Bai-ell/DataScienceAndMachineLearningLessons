FROZEN LAKE: Эпизоды и Цель обучения

Эпизод [1]

Первое действие в цепочке действий опишем следующим набором  s, a, r, s’:

https://lms-cdn.skillfactory.ru/assets/courseware/v1/f66da4163145b77de1a81f1eae74d090/asset-v1:SkillFactory+MLDL+ALWAYS+type@asset+block/DL_11_модуль_7_1.png

Продолжаем далее действия по предсказаниям политики на состояния (2, 0) → (2, 1) → (2, 2) → (2, 3) → (3, 3) до получения награды и окончания эпизода. Суммарная награда равна 10.

Эпизод — это некоторая игра, которую мы проиграли от начала и до конца. Размер эпизода может быть ограничен заданным количеством действий или состоянием среды. 

Эпизод [2]

Пусть во втором эпизоде политика подсказывает неверные действия и агент попадает в позицию (1, 1) ЯМА. Эпизод на этом завершается, суммарная награда равна 0.
