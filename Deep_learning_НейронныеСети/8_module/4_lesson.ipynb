 FROZEN LAKE: Состояния и действия

Разбирать обучение с подкреплением и вводить терминологию будем на простом примере игры «FROZEN LAKE».

Задача агента в игре дойти до ЦЕЛИ по квадратному полю 4 × 4, не попав в ЯМУ.

Состояния

В состояния будем вкладывать информацию о части окружающего мира, понимая под состоянием положение в конкретной ячейке поля. Имеем 16 состояний: s1, ..., s16. Тот факт, что робот находится в конкретной ячейке однозначно задаёт состояние системы.

Действия

Робот может совершать 4 действия: пойти вправо, влево, вверх, вниз. Имеем 4 действия a1, ..., a4. Важно понимать, что действие не всегда приводит к смене состояния. Часто действие стОит рассматривать как попытку действия. В нашем случае имеем пример дискретного списка действий и дискретный набор состояний.
