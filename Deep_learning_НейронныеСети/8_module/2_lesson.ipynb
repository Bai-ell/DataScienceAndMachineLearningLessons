Обучение с подкреплением

Пусть заданы:

Среда — это окружающий мир или симулятор игры, то есть то, что агент может наблюдать и с чем взаимодействовать.
Агент — это наш бот, модель или программа, которая в заданной среде совершает некоторые действия.
Агент может от среды получать наблюдения о состоянии среды (полное или частичное). Согласно своим наблюдениям агент может выполнять действия. По мере совершения действий в ответ агент может получить награду от среды. 

Задача: спланировать действия агента так, чтобы они привели к максимальной финальной награде.
