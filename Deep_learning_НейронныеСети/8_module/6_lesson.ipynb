FROZEN LAKE: s, a, r, s’

Рассмотрим, как агент может взаимодействовать со средой, и какие новые факторы и термины появляются.

Разберём буквенные обозначения s, a, r, s’:

s — состояние (state);
a — действие (action);
r —  награда (reward);
s’ —  следующее состояние, в которое попадёт агент, совершив действие.
Рассмотрим способ получения цепочки «s, a, r, s’»:

Среда возвращает агенту состояние.
Агент по этому состоянию принимает решение, какое действие совершить. Делает он это на основе политики. Политика —  это функция, которая ставит действие в соответствие состоянию. 
Агент сообщает среде о намерении совершить действие. 
Среда решает, какую дать награду за это действие и в какое следующее состояние перейдёт агент.
