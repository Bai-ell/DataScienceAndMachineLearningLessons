ONE-SHOT LEARNING


Допустим, мы имеем модель, обученную делать «хорошие» эмбеддинги, когда объекты одного класса находятся рядом, разных классов  —  далеко друг от друга.

Такая нейронная сеть обладает еще одним свойством.

После обучения можно подать в сеть объект нового класса, который она никогда не видела во время обучения. Первый «Инференс» с объектом нового класса назовём обучением, т.к. всю системы мы «обучаем» на новом классе по одной фотографии. Обучение сводится только лишь к инференсу через обученную эмбеддинг-модель и получению векторного представления попугая. Вектор, соответствующий новому классу, и является результатом обучения.

Если новый объект в режиме инференса попадёт в вектор, близкий по расстоянию к первому вектору, то этот объект будет классифицирован как «попугай». В случае близости с векторами других классов — классифицирован как «песик» или «котик».

Мы «смешали» задачу обучения с учителем, задачу кластеризации и др., используя таким образом их как One-shot learning.

Резюмируя: идея заключается в том, что новый класс задаётся векторным представлением одного экземпляра. Обучение проходит благодаря замечательному свойству «близкие объекты — близко, далёкие — далеко». Понятие класс стирается, остаётся понятие близости.

Обучение можно улучшить, проводя его по группе образцов.

Часто эту область знания называют metric-learning, т.е. «обучение расстоянию». Он часто используется в верификации лиц.
