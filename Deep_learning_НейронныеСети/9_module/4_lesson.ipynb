Сиамские нейронные сети

Сиамские нейронные сети позволяют решить проблему большого объёма данных, необходимого для обучения нейронной сети. Попытки решить эту проблему носят название «обучение по небольшому количеству прецедентов» (few-shot learning или one-shot learning).

К понятию ONE-SHOT LEARNING придём постепенно.

Обучение эмбеддингов

Вспомним, что эмбеддинг получается так: объект с помощью нейронной сети кодируется и представляется в виде вектора в некотором векторном пространстве. Задача состоит в том, чтобы эмбеддинг обучился таким образом, чтобы объекты одного класса в векторном пространстве он помещал рядом, а объекты разных классов — на бОльшем расстоянии. Иными словами, чтобы в этом эмбеддинговом пространстве объекты одного класса кластеризовались.

Сиамская нейронная сеть может решить поставленную задачу с помощью специальной функции Loss и позволяет подавать на вход несколько объектов, выдавая также несколько выходов. Это практически эквивалентно тому, если бы картинки подавались последовательно. Но здесь во время обучения учитываются оба объекта, и сеть в этом случае подобна двум разным сетям, но с одними и теми же весами.

Близкими объектами будем называть объекты из одного класса.

Для обучения такой нейронной сети на входе нужна обучающая информация или метки для разных классов картинок. Пусть такая обучающая выборка у нас есть. Сиамскую нейронную сеть будем обучать таким образом, чтобы картинки одного класса попадали в векторное пространство с векторами, находящимися на маленьком расстоянии d; а картинки из разных классов — на большом расстоянии.

Далее рассмотрим, как использовать эту модель для обучения по небольшому количеству образцов — one-shot learning.
