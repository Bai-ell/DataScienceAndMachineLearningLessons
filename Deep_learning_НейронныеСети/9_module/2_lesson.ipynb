Рассмотрим первый тип нейронных сетей — автоэнкодеры (или автокодиривщики).

В чем идея такой нейронной сети

Рассмотрим свёрточные автокодировщики. Подавая в нейронную сеть картинку, на выходе получим не классификацию, а скрытое промежуточное представление. Будем называть это признаковое представление кодом.

Автоэнкодер — это нейронная сеть, которая стоит на границе обучения без учителя и с учителем. Обучающей информации нет, т.к. код неизвестен. Но с другой стороны этот метод всё же будет обучением с учителем, потому что мы правильным образом сформируем таргет, к которому сеть будет стремиться. Для этого дополним сеть таким образом, чтобы на выходе получался такой же объект, какой был на входе. Именно так и будем обучать нейронную сеть.

https://lms-cdn.skillfactory.ru/assets/courseware/v1/5203eed742bc9b035c30cbb8e21fa6ca/asset-v1:SkillFactory+MLDL+ALWAYS+type@asset+block/DL-1_12_модуль_2.png

Таргет (целевой объект сети) — это наш входной объект. Таким образом учим сеть выдавать входной объект на выходе. Эта задача нетривиальная и важный момент в том, что код является маленьким компактным представлением, сжатой информацией (не без потерь). Задача — развернуть сжатую информацию из кода в исходной оригинал.

Название автокодировщик  обусловлено тем, что мы имеем дело с функцией, которая сама выучивает, как информацию сжать и как потом её разжать.

Обучение автоэнкодеров

Имеющиеся объекты используются как целевые объекты, к которым тянем сеть. Работая с картинками, можно сформировать функцию Loss, считая MSE-ошибку.

Такая нейронная сеть выглядит симметрично и очень похожа на случай сегментации.

Условно такую сеть можно разделить на Encoder и Decoder, обучая их сообща end-to-end.
