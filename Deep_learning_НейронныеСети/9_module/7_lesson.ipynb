Генеративные состязательные сети

Вариационные автоэнкодеры тоже являются генеративной моделью, но генеративные состязательные сети (англ. Generative adversarial network, сокращённо GAN)  обладают более качественным набором алгоритмов. 

Генератор состязательной сети получает на вход, например, случайный вектор, а на выходе синтезирует, например, картинку. Речь идёт о свёрточных GAN'ах. Информации о том, что необходимо синтезировать, нет. Это означает сэмплирование из некоторого пространства, например, пространства лиц. Таким образом, перед генератором мы ставим задачу выдать случайное лицо. Также здесь можно было бы поставить задачу сгенерировать картинку по тексту.

Итак, рассмотрим ситуацию со случайным вектором на входе. Рассматриваемый нами генератор обладает низким качеством. Улучшить его работу можно с помощью нейронной сети, которую назовём дискриминатор. Эта сеть должна будет выполнять оценку качества сгенерированной картинки, являясь по сути бинарным классификатором.



https://lms-cdn.skillfactory.ru/assets/courseware/v1/dd64c155e23847626fb98d51f668ecd4/asset-v1:SkillFactory+MLDL+ALWAYS+type@asset+block/DL-1_12_модуль_13.png

По картинке, полученной на вход, дискриминатор говорит, истинная (правильная) ли картинка получена или это синтезированная фейковая картинка. 

Задача дискриминатора — научиться отличать фейки от настоящих картинок, а задача генератора — обмануть дискриминатор, сделав так, чтобы полученную картинку дискриминатор оценивал как хорошую, т.е. реальную.

Эти две сети обучаются параллельно. В итоге обучения мы получаем идеальный генератор и идеальный дискриминатор.
