Введение. Продвинутые задачи

В предыдущих модулях мы рассматривали классические методы и задачи в глубоком обучении. Как правило, это были задачи распознавания или небольшой простой обработки. Была затронута нами также тема обучения с подкреплением.

В этом модуле поговорим о продвинутых темах, с которыми приходится сталкиваться в глубоком обучении. Рассмотрим существующие продвинутые задачи и методы.

Продвинутые задачи:

Понижение размерности данных.
Извлечение признаков.
Получение эмбеддингов.
Обучение по небольшому количеству прецедентов.
Генерация контента.
